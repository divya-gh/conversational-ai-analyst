## Understanding the meaning of each column


tweet_id- Unique ID for each tweet  
author_id- Who wrote the tweet (customer or company)  
inbound- True = customer message, False = company reply  
text- The actual tweet content  
created_at- Timestamp of the tweet   
response_tweet_id-  the tweet(s) that responded to it (could be multiple id's respondign to the tweet)  
in_reply_to_status_id- Which tweet this message is replying to (helps build conversation threads)

## Quick mental model (super helpful)
#### tweets like nodes in a tree:

```
tweet A
 ├── tweet B (reply)
 ├── tweet C (reply)
 │     └── tweet D (reply to C)
```
- response_tweet_id tells you children
- in_response_to_tweet_id tells you parent  
This is how we’ll build conversation IDs later.

## Missing values (your isna output)
We saw:  
response_tweet_id → ~1,040,629 missing

in_response_to_tweet_id → ~794,335 missing

This is normal. Beacause:
- Many tweets are root tweets (not replies)
- Many tweets have no replies
- Some conversations are one-way  
This is expected in customer support.
