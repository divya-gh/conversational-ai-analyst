### Missing intents in current taxonony:
Createa function to check for missing intent.   
##### Consider the following:
- return_request
- order_cancellation
- fraud_check
- account_access_issue
- technical_issue
- shipping_delay
- general_complaint
- product_support_issue

```python
def set_intent(text):
    if text == "" :
        return "non intent"
    text_l = text.lower()
    if "return" in  text_l or "faulty" in text_l :
        return  'return_request'
    if "scam" in text_l or "fraud" in text_l:
        return "fraud_check"
    if "login" in text_l or "account" in text_l or "reinstate" in text_l:
        return "account_access_issue"
    if "shipping" in text_l or "delay" in text_l or "tracking" in text_l:
        return "shipping_delay"
    if "cannot" in text_l or "not working" in text_l or "installed windows" in text_l:
        return "technical_issue"
    if "delivery" in text_l or "waiting" in text_l or "how long" in text_l:
        return "delivery_status"
    return "unknown_candidate" 

df_unlnown_text["suggested_intent"] = df_unlnown_text["Clean_text"].fillna("").apply(set_intent)
df_unlnown_text[["Clean_text", "suggested_intent"]].head(20)
```

### Analyze if text is more of a billing or refund issue
Example: 
1. i am sorry for the condition your package arri...	billing_issue  
- is more of a refund_request

2.3 Quantify overlap by looking at n‑grams:

```python
from collections import Counter

def top_ngrams(texts, n=2, top_k=20):
    all_ngrams = []
    for t in texts:
        tokens = t.lower().split()
        all_ngrams.extend([" ".join(tokens[i:i+n]) for i in range(len(tokens)-n+1)])
    return Counter(all_ngrams).most_common(top_k)

billing_texts = df[df["intent"] == "billing_issue"]["Clean_text"].dropna().tolist()
refund_texts = df[df["intent"] == "refund_request"]["Clean_text"].dropna().tolist()

top_billing = top_ngrams(billing_texts, n=2)
top_refund = top_ngrams(refund_texts, n=2)
top_billing_refund = pd.DataFrame({'top_billing': top_billing,
                                   'top_refund': top_refund
                                  })
top_billing_refund
```
